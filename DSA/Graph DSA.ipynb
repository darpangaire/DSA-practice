```python
# time complexity 0(V+E)
# Space complexity O(V)
# checking bipartite using BFS
from collections import deque
def is_bipartite(graph):
    n = len(graph)
    color = [-1]*n
    
    for start in range(n):
        if color[start] == -1:
            color[start] = 0
            queue = deque([start])
            
            while queue:
                u = queue.popleft()
                for v in graph[u]:
                    if color[v] == -1:
                        color[v] = 1-color[u]
                        queue.append(v)
                        
                    elif color[u] == color[v]:
                        return False
                    
    return True

graph = [
    [1],
    [0,2,3],
    [1],
    [1,4,5,8],
    [3],
    [3,6],
    [5,7],
    [6,8],
    [7,3]
]
        
is_bipartite(graph)
```

```python
# prerequisite task
# kahn's algorithm
from collections import defaultdict, deque
def is_possible(prerequisities,n,p):
    adj = defaultdict(list)
    in_degree = [0]*n
    for u,v in prerequisities:
        adj[v].append(u)
        in_degree[u] += 1
        
    # kahn's algorithm
    queue = deque()
    count = 0
    for i in range(n):
        if in_degree[i] == 0:
            queue.append(i)
            
    while queue:
        node = queue.popleft()
        count += 1
        for neighbour in adj[node]:
            in_degree[neighbour] -= 1
            if in_degree[neighbour] == 0:
                queue.append(neighbour)
                
    return count == n

pre = [(1,0),(2,1),(3,2)]
N = 4
P = 3
is_possible(pre,N,P)      
```

```python
# ALien Dictionary
from collections import deque,defaultdict

def Aliens_algorithm(st,t):
    n = len(st)
    adj = defaultdict(list)
    indegree = [0]*t
    for i in range(n-1):
        str1 = st[i]
        str2 = st[i+1]
        j,k = 0,0
        while j < len(str1) and k < len(str2) and str1[j] == str2[k]:
            j += 1
            k += 1
        
        if j == len(str1) or k == len(str2):
            continue
        adj[ord(str1[j])-97].append(ord(str2[k])-97)
        indegree[ord(str1[j])-97] += 1
        
    # kahn's algorithm
    queue = deque()
    count = 0
    for i in range(t):
        if indegree[i] == 0:
            queue.append(i)
            
    while queue:
        node = queue.popleft()
        count += 1
        for neighbour in adj[node]:
            indegree[neighbour] -= 1
            if indegree[neighbour] == 0:
                queue.append(neighbour)
                
    return count == k
        

    
st = ['baa','abcd','abca','cab','cad']
Aliens_algorithm(st,4)
```

```python
#Alien dictionary
from collections import defaultdict,deque
def alienorder(words):
    n = len(words)
    adj = defaultdict(set)
    indegree = {char:0 for word in words for char in word}
    for i in range(n-1):
        st1 = words[i]
        st2 = words[i+1]
        min_len = min(len(st1),len(st2))
        for j in range(min_len):
            if st1[j] != st2[j]:
                if st2[j] not in adj[st1[j]]:
                    adj[st1[j]].add(st2[j])
                    indegree[st2[j]] += 1
                    
                break
        else:
            if len(st1) > len(st2):
                return ''
            
    queue = deque([char for char in indegree if indegree[char] == 0])
    sorted_char = []
    
    while queue:
        char = queue.popleft()
        sorted_char.append(char)
        for neighbour in adj[char]:
            indegree[neighbour] -= 1
            if indegree[neighbour] == 0:
                queue.append(neighbour)
                
    if len(sorted_char) == len(indegree):
        return sorted_char
    
    else:
        return ' '
    
words = ['wrt','wrf','er','ett','rftt']
print(alienorder(words))
```

```python
from collections import defaultdict,deque
def maxi_time(relation,time,n):
    adj = defaultdict(list)
    indegree = [0]*n
    for u,v in relation:
        adj[u-1].append(v-1)
        indegree[v-1] += 1
        
    queue = deque()
    for i in range(n):
        if indegree[i] == 0:
            queue.append(i)
    
    course_time = [0]*n
    for i in range(n):
        course_time[i] = time[i]
    
    while queue:
        node = queue.popleft()
        for neighbor in adj[node]:
            course_time[neighbor] = max(course_time[neighbor],course_time[node] + time[neighbor])
            indegree[neighbor] -= 1
            if indegree[neighbor] == 0:
                queue.append(neighbor)
                
    return max(course_time)

relations = [[1,5],[2,5],[3,5],[3,4],[4,5]]
time = [1,2,3,4,5]
maxi_time(relations,time,5)  
```

```python
# Dijkstra's algorithm
# implemenatation of dijkstra algorithm
# time complexity E log v
# worst time complexity O(V^2 log v)
import heapq
def dijkstra_algorithm(n,adj,s):
    dist = [float('inf')]*n
    dist[s] = 0
    priority_queue = [(0,s)]
    while priority_queue:
        d,u = heapq.heappop(priority_queue)
        if d > dist[u]:
            continue
            
        for v,wt in adj[u]:
            if dist[u] + wt < dist[v]:
                dist[v] = dist[u]+wt
                heapq.heappush(priority_queue,(dist[v],v))
                
                
    return dist
                

V = 3
adj = [[[1, 1], [2, 6]], [[2, 3], [0, 1]], [[1, 3], [0, 6]]]
S = 2
dijkstra_algorithm(V,adj,S)
```

```python
# time complexity O(V*E)
# Space complexity O(V)
def bellman_ford(graph,n,src):
    dist = [float('inf')]*n
    dist[src] = 0
    for _ in range(n-1):
        for u,v,wt in graph:
            if dist[u] != float('inf') and dist[u] + wt < dist[v]:
                dist[v] = dist[u] + wt
                
    for u,v,wt in graph:
        if dist[u] != float('inf') and dist[u] + wt < dist[v]:
            print('graph contains negative cycle')
            return
        
    return dist
        

graph = [
    [0, 1, -1],
    [0, 2, 4],
    [1, 2, 3],
    [1, 3, 2],
    [1, 4, 2],
    [3, 2, 5],
    [3, 1, 1],
    [4, 3, -3]
]

print(bellman_ford(graph,5,0))
    
```

```python
# Shortest Source to Destination Path in Binary Matrix
# time complexity O(N*M)
# space complexity O(min(N,M))
from collections import deque
def shortest_src(mat,x,y):
    queue = deque([(0,0,0)])
    global r,c
    r = len(mat)
    c = len(mat[0])
    row = [1,-1,0,0]
    col = [0,0,-1,1]
                
    while queue:
        u,v,dist = queue.popleft()
        if check(u,v,x,y):
            return dist
        for k in range(4):
            if valid(row[k]+u,col[k]+v) and mat[row[k]+u][col[k]+v] == 1:
                mat[row[k]+u][col[k]+v] = 0
                queue.append((row[k]+u,col[k]+v,dist+1))

    return -1
              
def valid(i,j):
    return i >= 0 and i < r and j >= 0 and j < c

def check(u,v,x,y):
    return x == u and y ==v

mat = [
    [1,0,0,0],
    [1,1,0,1],
    [0,1,1,1]
]
shortest_src(mat,2,3)
```

```python
# minimum spanning tree
# time complexity O(V+E)logV for list
# time complexity 0(v^2) for matrix
# SPACE COMPLEXITY O(V)
# prim's algorithm
import heapq

def is_mst(adj,n):
    ismst = [0]*n
    parent = [None]*n
    min_heap = []
    cost = 0
    
    heapq.heappush(min_heap,(0,0)) # (weight,node)
    
    while min_heap:
        wt,node = heapq.heappop(min_heap) 
        if not ismst[node]:
            ismst[node] = 1
            cost += wt
            
            for neighbor,weight in adj[node]:
                heapq.heappush(min_heap,(weight,neighbor))
                parent[neighbor] = node
                
    return cost


adj = {
    0: [(1, 2), (3, 6)],
    1: [(0, 2), (2, 3), (3, 8), (4, 5)],
    2: [(1, 3), (4, 7)],
    3: [(0, 6), (1, 8)],
    4: [(1, 5), (2, 7)]
}
n = 5 

is_mst(adj,n)
    
    
    
```

```python
# bridge of a graph
from collections import defaultdict

class Graph:
    def __init__(self,vertices):
        self.v = vertices
        self.graph = defaultdict(list)
        self.time = 0
        
    def add_edge(self,u,v):
        self.graph[u].append(v)
        self.graph[v].append(u)
        
    def bridges_util(self,u,visited,parent,dist,low):
        visited[u] = True
        dist[u] = self.time
        low[u] = self.time
        self.time += 1
        
        for v in self.graph[u]:
            if not visited[v]:
                parent[v] = u
                self.bridges_util(v,visited,parent,dist,low)
                
                low[u] = min(low[u],low[v])
                if dist[u] < low[v]:
                    print("bridge vertex: ",u,"->",v)
                    
            elif v != parent[u]:
                low[u] = min(low[u],dist[v])
    
    def find_bridges(self):
        visited = [False]*self.v
        parent = [-1]*self.v
        dist = [float('inf')]*self.v
        low = [float('inf')]*self.v
        
        for i in range(self.v):
            if not visited[i]:
                self.bridges_util(i,visited,parent,dist,low)
                
                
g = Graph(5)
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(1, 3)
g.add_edge(3, 4)

g.find_bridges()       
                
```

```python
# tarjan's algorithm
# articulation point

def articulation_point(adj,v):
    visited = [False]*v
    parent =[-1]*v
    disc = [float('inf')]*v
    low = [float('i  nf')]*v
    ans = set()
    timer = [0]
    
    def dfs(node):
        children = 0
        visited[node] = True
        disc[node] = low[node] = timer[0]
        timer[0] += 1
        
        for neighbor in adj[node]:
            if not visited[neighbor]:
                children += 1

                parent[neighbor] = node
                dfs(neighbor)
                low[node] = min(low[node],low[neighbor])

                if parent[node] != -1 and low[neighbor] >= disc[node]:
                    ans.add(node)

                if parent[node] == -1 and children > 1:
                    ans.add(node)
                    
            elif neighbor != parent[node]:
                low[node] = min(low[node],disc[neighbor])
                
    for i in range(v):
        if not visited[i]:
            dfs(i)
            
    return list(ans)

# Example usage:
# Graph is represented as an adjacency list
adj_list = [
    [1, 2],    # 0
    [0, 2],    # 1
    [0, 1, 3, 5],  # 2
    [2, 4],    # 3
    [3],       # 4
    [2, 6, 8], # 5
    [5, 7],    # 6
    [6],       # 7
    [5, 9, 10],# 8
    [8, 10],   # 9
    [8, 9]     # 10
]

v = len(adj_list)
print(articulation_point(adj_list,v))
```
