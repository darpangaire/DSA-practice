```python
# check whether the binary tree is heap or not.
# time complexity is O(n)
# space complexity is O(logn)
class Heap:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
    def count_nodes(self,root):
        if root is None:
            return 0
        return 1+ self.count_nodes(root.left) + self.count_nodes(root.right)
    
    def heap_property_utils(self,root):
        if root.left is None and root.right is None:
            return True
        
        if root.right is None:
            return root.data >= root.left.data
        
        else:
            if root.left.data <= root.data >= root.right.data:
                return self.heap_property_utils(root.left) and self.heap_property_utils(root.right)
            
            else:
                return False
            
    def complete_tree_util(self,root,index,node_count):
        if root is None:
            return True
        
        if index >= node_count:
            return False
        
        return (self.complete_tree_util(self,2*index+1,node_count) and 
               self.complete_tree_util(self,2*index+2,node_count))
    
    def check_if_heap(self):
        node_count = self.count_nodes(self)
        if (self.complete_tree_util(self,0,node_count) and
           self.heap_property_utils(self)):
            return True
        
        else:
            return False
        
        
heap = Heap(5)
heap.left = Heap(4)
heap.right = Heap(3)
heap.left.left = Heap(2)
heap.left.right = Heap(1)

if heap.check_if_heap():
    print("Given Tree is Binary")
    
else:
    print("Given Tree is not Binary")
        
```

```python
# Kth smallest element in sorted matrix
def count_less_equal(matrix,mid):
    n = len(matrix)
    count = 0
    row, column = n-1,0
    while row >= 0 and column < n:
        if matrix[row][column] <= mid:
            count += (row+1)
            column += 1
            
        else:
            row -= 1
    return count
            

def Kth_smallest_element(matrix,k):
    n = len(matrix)
    left,right = matrix[0][0],matrix[n-1][n-1]
    while left < right:
        mid = (left + right)//2
        if count_less_equal(matrix,mid) < k:
            left = mid + 1
            
        else:
            right = mid
            
    return left

mat = [[16,28,60,64],
      [22,41,63,91],
      [27,50,87,93],
      [36,78,87,94]]
k = 6

Kth_smallest_element(mat,k)

```

```python
import heapq

class MidFinder:
    def __init__(self):
        self.low = []
        self.high = []
        
    def AddNumber(self,num):
        heapq.heappush(self.low,-num)
        if self.low and self.high and (-self.low[0] > self.high[0]):
            heapq.heappush(self.high,-heapq.heappop(self.low))
            
        if len(self.low) > len(self.high)+1:
            heapq.heappush(self.high,-heapq.heappop(self.low))
            
        if len(self.high) > len(self.low):
            heapq.heappush(self.low,-heapq.heappop(self.high))
            
    def findMedian(self):
        if len(self.low) > len(self.high):
            return - self.low[0]
        
        return(-self.low[0] + self.high[0])//2
    
medianFinder = MidFinder()
num = [41,35,62,4,97,108]
for number in num:
    medianFinder.AddNumber(number)
    print("Current median: ",medianFinder.findMedian())
```

```python
# merge k sorted matrix
#Time complexity O(Nlogk)
#space complexity O(k)
import heapq
def merge_k_sorted_mat(mat):
    min_heap = []
    for i in range(len(mat)):
        if mat[i]:
            heapq.heappush(min_heap,(mat[i][0],i,0))
        
    result = []
    #tc O(Nlogk)
    while min_heap:
        value,mat_indx,ele_indx = heapq.heappop(min_heap)
        result.append(value)
        if ele_indx+1 < len(mat):
            next_value = mat[mat_indx][ele_indx+1]
            # this teps take log k
            heapq.heappush(min_heap,(next_value,mat_indx,ele_indx+1))
            
    return result

mat = [
    [1,5,9],
    [2,6,8],
    [3,7,10]
]

merge_k_sorted_mat(mat)  
```

```python
# Smallest Range Covering Elements from K Lists
import heapq
class Solution:
    def smallestRange(self,nums,k):
        pq = []
        mmax = 0
        for i in range(k):
            heapq.heappush(pq,(nums[i][0],i,0))
            mmax = max(mmax,nums[i][0])
            
        ans = [pq[0][0],mmax]
        while True:
            _,mat_indx,ele_indx = heapq.heappop(pq)
            if ele_indx == len(nums[mat_indx])-1:
                break
                
            next_num = nums[mat_indx][ele_indx+1]
            mmax = max(mmax,next_num)
            heapq.heappush(pq,(next_num,mat_indx,ele_indx+1))
            
            if mmax - pq[0][0] < ans[1] - ans[0]:
                ans = [pq[0][0],mmax]
                
        return ans
         
    
arr = [
        [4, 7, 9, 12, 15],
        [0, 8, 10, 14, 20],
        [6, 12, 16, 30, 50]
    ]

k = 3

s = Solution()
s.smallestRange(arr,3)
```
