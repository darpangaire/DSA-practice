```python
# Time complexity 0(NlogN)
# Space complexity O(N).
from collections import defaultdict,deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
def vertical_traversal(root):
    if not root:
        return None
    
    column_table = defaultdict(list)
    queue = deque([(root,0,0)])
    
    while queue:
        node,row,col = queue.popleft()
        if node:
            column_table[col].append((row,node.val))
            queue.append((node.left,row+1,col-1))
            queue.append((node.right,row+1,col+1))
            
    sorted(column_table.items())
    result = []
    for column, value in sorted(column_table.items()):
        result.append([val for row,val in sorted(value)])
    
    return result


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("Vertical Traversal of the Binary Tree:")
print(vertical_traversal(root))
```

```python
#  Diagonal Traversal
from collections import defaultdict,deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
def diagonal_traversal(root):
    if not root:
        return []
    
    diagonal_table = defaultdict(list)
    queue = deque([(root,0)])
    
    while queue:
        node,diagonal = queue.popleft()
        
        while node:
            diagonal_table[diagonal].append(node.val)
            if node.left:
                queue.append((node.left,diagonal+1))
                
            node = node.right
            
    result = []
    for diagonal in sorted(diagonal_table.keys()):
        result.append(diagonal_table[diagonal])
        
    return result


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("Diagonal Traversal of the Binary Tree:")
print(diagonal_traversal(root))
    
```

```python
# boundary traversal of binary tree
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
def is_Leaf(node):
    return node.left is None and node.right is None

def add_left_boundary(node,boundary_nodes):
    while node:
        if not is_Leaf(node):
            boundary_nodes.append(node.val)
            
        if node.left:
            node = node.left
            
        else:
            node = node.right
            
def add_right_boundary(node,boundary_nodes):
    stack = []
    while node:
        if not is_Leaf(node):
            stack.append(node.val)
            
        if node.right:
            node = node.right
            
        else:
            node = node.left
            
    while stack:
        boundary_nodes.append(stack.pop())
        
def add_leaves(node,boundary_nodes):
    if node != None:
        add_leaves(node.left,boundary_nodes)
        if is_Leaf(node):
            boundary_nodes.append(node.val)
        add_leaves(node.right,boundary_nodes)
        
def boundary_traversal(root):
    if not root:
        return []
    
    boundary_nodes = []
    
    if not is_Leaf(root):
        boundary_nodes.append(root.val)
        
    add_left_boundary(root.left,boundary_nodes)
    add_leaves(root,boundary_nodes)
    add_right_boundary(root.right,boundary_nodes)
    
    return boundary_nodes


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print("Diagonal Traversal of the Binary Tree:")
print(diagonal_traversal(root))
```
